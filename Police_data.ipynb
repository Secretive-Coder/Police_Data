```python
#Que2. Question ( Based on Filtering + Value Counts ) - For Speeding , were Men or Women stopped more often ?
speeding_data = data[data['violation'] == 'Speeding']
gender_counts = speeding_data['driver_gender'].value_counts()

print("Number of stops for speeding by gender:")
print(gender_counts)
```

```python
#Que3. Question ( Groupby ) - Does gender affect who gets searched during a stop ?
gb = data.groupby("driver_gender")
searches_by_gender = gb.size()
print(searches_by_gender)
```

```python
#Que4. Question ( mapping + data-type casting ) - What is the mean stop_duration ?
mean_value = pd.to_numeric(data['stop_duration'], errors='coerce').mean()
print("Mean value of stop duration:", mean_value)
```

```python
#Que5. Question ( Groupby , Describe ) - Compare the age distributions for each violation.
import pandas as pd

data = pd.read_csv(r"C:\Users\jangh\Downloads\Police Data.csv")

data['driver_age'] = pd.to_numeric(data['driver_age'], errors='coerce')
data['violation'] = pd.Categorical(data['violation'])

grouped = data.groupby('violation')['driver_age'].describe()

print(grouped)
```
